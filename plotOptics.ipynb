```python
#!/usr/bin/env python3
import argparse
import uproot
import matplotlib.pyplot as plt
import numpy as np

fin = uproot.open("optics.root")
tin = fin["Optics"]
s = tin["S"].array()
npart = tin["Npart"].array()
beta_x = tin["Beta_x"].array()
beta_x_sigma = tin["Sigma_Beta_x"].array()
beta_y = tin["Beta_y"].array()
beta_y_sigma = tin["Sigma_Beta_y"].array()
emt_x = tin["Emitt_x"].array()
emt_x_sigma = tin["Sigma_Emitt_x"].array()
emt_y = tin["Emitt_y"].array()
emt_y_sigma = tin["Sigma_Emitt_y"].array()
alp_x = tin["Alpha_x"].array()
alp_x_sigma = tin["Sigma_Alpha_x"].array()
alp_y = tin["Alpha_y"].array()
alp_y_sigma = tin["Sigma_Alpha_y"].array()
mean_x = tin["Mean_x"].array()
mean_x_sigma = tin["Sigma_Mean_x"].array()
mean_y = tin["Mean_y"].array()
mean_y_sigma = tin["Sigma_Mean_y"].array()

tobj = fin["Model"]
comp_type = tobj["Model./Model.componentType"].array()
comp_name = tobj["Model./Model.componentName"].array()
comp_plac = tobj["Model./Model.placementName"].array()
comp_leng = tobj["Model./Model.length"].array()
print ( len(comp_leng[0]) )
print ( len(comp_plac[0]) )
print ( len(npart) )

acc_len=0
for i in range(len(comp_type[0])):
    acc_len = acc_len+comp_leng[0][i]
    print(comp_plac[0][i], comp_type[0][i], comp_name[0][i], acc_len, int(npart[i]))
    if i > 100:
        break



```

```python
plt.scatter(s, npart, label="Npart vs S", alpha=0.7)
plt.xlabel("S (m)")
plt.ylabel("N particles")
plt.legend()
plt.show()

plt.errorbar(s, mean_x, yerr=mean_x_sigma, fmt='o', label="Mean X vs S", alpha=0.7)
plt.xlabel("S (m)")
plt.ylabel("Mean X (m)")
plt.text(s[-1], mean_x[-1], f"({mean_x[-1]:.2e})", fontsize=9, color='red')
plt.text(s[0], mean_x[0], f"({mean_x[0]:.2e})", fontsize=9, color='red')
plt.legend()
plt.show()

plt.errorbar(s, mean_y, yerr=mean_y_sigma, fmt='o', label="Mean Y vs S", alpha=0.7)
plt.xlabel("S (m)")
plt.ylabel("Mean Y (m)")
plt.text(s[-1], mean_y[-1], f"({mean_y[-1]:.2e})", fontsize=9, color='red')
plt.text(s[0], mean_y[0], f"({mean_y[0]:.2e})", fontsize=9, color='red')
plt.legend()
plt.show()

plt.errorbar(s, beta_x, yerr=beta_x_sigma, fmt='o', label="Beta X vs S", alpha=0.7)
plt.xlabel("S (m)")
plt.ylabel("Beta X (m)")
plt.text(s[-1], beta_x[-1], f"({beta_x[-1]:.2f})", fontsize=9, color='red')
plt.text(s[0], beta_x[0], f"({beta_x[0]:.2f})", fontsize=9, color='red')
plt.legend()
plt.show()

plt.errorbar(s, beta_y, yerr=beta_y_sigma, fmt='o', label="Beta Y vs S", alpha=0.7)
plt.xlabel("S (m)")
plt.ylabel("Beta Y (m)")
plt.text(s[-1], beta_y[-1], f"({beta_y[-1]:.2f})", fontsize=9, color='red')
plt.text(s[0], beta_y[0], f"({beta_y[0]:.2f})", fontsize=9, color='red')
plt.legend()
plt.show()

plt.errorbar(s, emt_x, yerr=emt_x_sigma, fmt='o', label="Emittance X vs S")
plt.xlabel("S (m)")
plt.ylabel("Emittance X (m)")
plt.text(s[-1], emt_x[-1], f"({emt_x[-1]:.2e})", fontsize=9, color='red')
plt.text(s[0], emt_x[0], f"({emt_x[0]:.2e})", fontsize=9, color='red')
plt.legend()
plt.show()

plt.errorbar(s, emt_y, yerr=emt_y_sigma, fmt='o', label="Emittance Y vs S")
plt.xlabel("S (m)")
plt.ylabel("Emittance Y (m)")
plt.text(s[-1], emt_y[-1], f"({emt_y[-1]:.2e})", fontsize=9, color='red')
plt.text(s[0], emt_y[0], f"({emt_y[0]:.2e})", fontsize=9, color='red')
plt.legend()
plt.show()
```
